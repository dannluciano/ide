# For - Para

```python
# para cada valor do intervalo de 0 até 10
for contador in range(10):
    print(contador)
```

```python
for _ in range(5):
    print("Ola Mundo")
```

## Underline _

```python
for _ in range(5):
    print("Ola Mundo", _)
```

```python
for _ in range(5):
    print("Ola Mundo")

print(_)
```

```python
print(_)
for _ in range(5):
    print("Ola Mundo")

print(_)
```
