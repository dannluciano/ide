```python
def exibe_menu():
    print("|==============================|")
    print("|  =*=  Sistema de Loja  =*=   |")
    print("+------------------------------+")
    print("| 1 - Adicionar um Produto     |")
    print("| 2 - Buscar um Produto        |")
    print("| 3 - Listar todos os Produtos |")
    print("| 4 - Remover um Produto       |")
    print("| 0 - Sair                     |")
    print("|==============================|")

def adiciona_produto():
    print("Adicionar um Produto.....")
    nome = input("Digite o nome do produto")
    preco = input("Digite o preço do produto")

opcao = 1
while opcao != "0":
    exibe_menu()
    print("Digite uma Opção")
    opcao = input()

    if opcao == "1":
        adiciona_produto()

print("Saindo...")
```

```python
def calcula_preco_com_desconto(preco, desconto):
    return preco-preco*desconto

preco = float(input("Digite o preço do produto: "))
desconto = int(input("Digite o desconto do produto: (1% - 100%)"))
calcula_preco_com_desconto(preco, desconto/100)
```

```python
def area_quadrado(lado):
    return lado * lado
```

```python
def area_retangulo(base, altura):
    return base * altura
```

```python
def area_trapezio(base_maior, base_menor, altura):
    return ((base_maior + base_menor) * altura) / 2
```
