# Loader

```python
import ipywidgets as widgets
seconds_slide = widgets.IntSlider(min=10, max=100)
seconds_slide
```

```python
import sys
import time

cols = 30
sleep_time = 0.1
steps = 100
total_time = 100

barracompleta = "[=> ]"
# barracompleta = "(=>*)"
# barracompleta = "|░▏ |"
barracompleta = "|█▏ |"

barrainicio = barracompleta[0]
barrafim = barracompleta[-1]
barra = barracompleta[1]
barrafinal = barracompleta[2]
espaco = barracompleta[3]

# loading = "-\\|/"
# loading = "x+"
# loading = "◜◠◝◞◡◟"
# loading = "◴◷◶◵"
# loading = " ▏▎▍▌▊█"
# loading = " ▏▎▍▌▊█▊▌▍▎▏"
# loading = "▁▂▃▄▅▆▇█"
loading = "▁▂▃▄▅▆▇█▇▆▅▄▃▂"

if len(sys.argv) >= 2:
    total_time = int(sys.argv[1])
    steps = int(total_time * 1 / sleep_time)
else:
    total_time = int(seconds_slide.value)
    steps = int(total_time * 1 / sleep_time)

for s in range(steps + 1):
    percent = s / steps * cols - 1
    real_percent = s / steps * 100
    print(f"\r{real_percent:6.2f} % {barrainicio}", end="")
    i = 0
    while i < percent and i < cols:
        i += 1
        print(barra, end="")

    print(barrafinal, end="")

    i += 1

    while i > percent and i < cols:
        i += 1
        print(espaco, end="")

    char = loading[s % len(loading)]
    eta = int(total_time - s * sleep_time)
    print(f"{barrafim} {char} ETA {eta:5}s", end="", flush=True)
    time.sleep(sleep_time)

print("\nFinished!")
```
