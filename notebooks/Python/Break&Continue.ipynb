# Break e Continue

## Break

Para a repetição

```python
while True:
    num = int(input(
        "Digite um Numero (Zero para sair): "
    ))
    if num == 0:
        break
```

```python
num = 1
while num != 0:
    num = int(input(
        "Digite um Numero (Zero para sair): "
    ))
    
```

```python
for i in range(1, 10, 1):
    if i == 5:
        break
    print(i)
```

## Continue

Vou para proxima repetição

```python
for i in range(1, 10, 1):
    if i % 2 == 0:
        continue
    print(i)
```

```python
while True:
    print("Comando1")
    print("Comando2")
    continue
    print("Comando3")
    print("Comando4")
```
